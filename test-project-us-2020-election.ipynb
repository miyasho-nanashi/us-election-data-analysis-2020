# a. Correlation Analysis
The relationship between the percentage of Republican votes and the racial demographics:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Correlation matrix
correlation = df[['2020 Republican vote %', 'NH-White percentage', 
                  'NH-Black percentage', 'NH-Asian percentage', 
                  'Hispanic or Latino percentage']].corr()

# Heatmap
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title('Correlation Between Race and Republican Vote %')
plt.show()
```

# Key Observations from the Heatmap

2020 Republican Vote % vs. NH-White Percentage:

Correlation: 0.54 (moderately strong positive correlation).
Interpretation: Counties with a higher percentage of non-Hispanic white populations tend to vote more Republican.

2020 Republican Vote % vs. NH-Black Percentage:

Correlation: -0.43 (moderate negative correlation).
Interpretation: Counties with a higher percentage of Black populations tend to vote less Republican.

2020 Republican Vote % vs. NH-Asian Percentage:

Correlation: -0.46 (moderate negative correlation).
Interpretation: Counties with a higher percentage of Asian populations tend to vote less Republican.

2020 Republican Vote % vs. Hispanic or Latino Percentage:

Correlation: -0.14 (weak negative correlation).
Interpretation: Hispanic or Latino populations have a very weak negative correlation with Republican votes, suggesting weaker trends compared to other racial groups.
Other Relationships:

NH-White Percentage vs. NH-Black Percentage: Correlation: -0.59. This suggests counties with higher white populations tend to have lower Black populations, which is expected due to demographic distribution.

Hispanic or Latino Percentage vs. NH-White Percentage: Correlation: -0.61. Counties with higher Hispanic populations tend to have lower white populations.

# What Does This Tell Us?

The heatmap indicates:

There is a moderate positive correlation between Republican voting and non-Hispanic white population percentage.
Conversely, there are moderate negative correlations between Republican voting and other racial demographics (Black, Asian, Hispanic).
This supports the idea that Republican support might be stronger in predominantly white counties, while Democratic support might be stronger in more diverse counties.

# b. Grouped Analysis
Group counties by racial majority and compare average Republican vote percentages:

```python
import numpy as np

# Add a column for racial majority
def racial_majority(row):
    races = ['NH-White percentage', 'NH-Black percentage', 'NH-Asian percentage', 'Hispanic or Latino percentage']
    majority_race = max(races, key=lambda x: row[x])
    return majority_race

df['racial_majority'] = df.apply(racial_majority, axis=1)

# Group by racial majority and calculate mean Republican vote %
grouped = df.groupby('racial_majority')['2020 Republican vote %'].mean()
print(grouped)

# Bar plot
grouped.plot(kind='bar', color='skyblue')
plt.title('Republican Vote % by Racial Majority in Counties')
plt.ylabel('Average Republican Vote %')
plt.xlabel('Racial Majority')
plt.show()
```

# Analysis highlights a key insight: 
while the stereotype may hold for some groups (e.g., strong Republican support in predominantly white counties), the high Republican support in Hispanic-majority counties challenges the simplistic narrative that GOP rhetoric universally alienates Hispanic voters.

# Key Observations from the Chart

# White Majority Counties (NH-White Percentage):

GOP support is the highest here (66.42%), which aligns with the traditional perception of the Republican voter base.

# Hispanic Majority Counties:

GOP support is surprisingly strong (56.74%) despite media narratives around immigration policies. This could reflect factors like:
Economic policies resonating with certain Hispanic groups.
Socially conservative values prevalent in some Hispanic communities.
Regional influences (e.g., Cuban Americans in Florida).
Asian and Black Majority Counties:

Republican support in these counties is much lower (34.98% and 35.37%, respectively), reflecting the broader trends of minority groups leaning Democratic.

# Complex Dynamics:

# c. Statistical Testing
Perform a t-test or ANOVA to check if differences in Republican vote % among racial groups are statistically significant:

```python
from scipy.stats import f_oneway

# Republican vote percentages for each racial-majority group
white_votes = df[df['racial_majority'] == 'NH-White percentage']['2020 Republican vote %']
black_votes = df[df['racial_majority'] == 'NH-Black percentage']['2020 Republican vote %']
asian_votes = df[df['racial_majority'] == 'NH-Asian percentage']['2020 Republican vote %']
hispanic_votes = df[df['racial_majority'] == 'Hispanic or Latino percentage']['2020 Republican vote %']

# Perform ANOVA test
stat, p_value = f_oneway(white_votes, black_votes, asian_votes, hispanic_votes)

# Results
print(f'ANOVA Statistic: {stat}, p-value: {p_value}')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Box plot for visualizing differences
sns.boxplot(data=df, x='racial_majority', y='2020 Republican vote %')
plt.title('Republican Vote % by Racial Majority')
plt.xlabel('Racial Majority')
plt.ylabel('Republican Vote %')
plt.xticks(rotation=45)
plt.show()
```

p-value is incredibly small (1.9157e-110), not greater than 0.05. 
This result indicates that the differences in Republican vote percentages across racial-majority groups are statistically significant.

What Does This p-Value Mean?
p-value < 0.05: Reject the null hypothesis.
This means that the differences in Republican vote percentages among racial-majority groups are not due to random chance. There is a statistically significant difference in GOP support based on the racial majority of counties.
The ANOVA test result (F-statistic: 184.97, p-value: ~0) suggests that at least one group’s Republican vote percentage differs significantly from the others.

# Tukey’s HSD Test
Tukey's test will identify which pairs of groups have statistically significant differences.

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd

# Perform Tukey's HSD test
tukey = pairwise_tukeyhsd(endog=df['2020 Republican vote %'], 
                          groups=df['racial_majority'], 
                          alpha=0.05)

# Print results
print(tukey)

# Optional: Visualize results
tukey.plot_simultaneous()
plt.title('Tukey HSD Test for GOP Support Across Racial Groups')
plt.show()
```

# Black and Asian Counties Show Low Support:
